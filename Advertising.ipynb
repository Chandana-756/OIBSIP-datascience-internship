```python
X = adv.drop('Sales',axis=1)
X
```

```python
data1 = X.drop(columns='Unnamed: 0')
data1
```

```python
data2 = adv[['Sales']]
data2
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test = train_test_split(data1,data2,test_size = 0.20, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
p = LinearRegression()
p.fit(X_train,Y_train)
```

```python
Y_predict = p.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print("Mean Absolute Error is ",metrics.mean_absolute_error(Y_predict,Y_test))
```

```python
print("Root Mean Squared Error is ",metrics.mean_squared_error(Y_predict,Y_test))
```

```python
print("R squared value is ", metrics.r2_score(Y_predict,Y_test))
```

```python
print("Maximum error is", metrics.max_error(Y_predict,Y_test))
```
